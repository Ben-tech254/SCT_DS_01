```python
#Importing all the necessary libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the dataset
t_data=sns.load_dataset('titanic')
t_data.head()
```

```python
#Information of the dataset
t_data.info()
```

```python
#Check for missing values
t_data.isna().sum()
```

```python
#Filling in missing values
t_data['age'].fillna(t_data['age'].mean(), inplace=True)
t_data['embarked'].fillna(t_data['embarked'].mode()[0], inplace=True)
t_data['deck'].fillna(t_data['deck'].mode()[0], inplace=True)
t_data['embark_town'].fillna(t_data['embark_town'].mode()[0], inplace=True)
t_data.isna().sum()
```

```python
#Plotting a histogram
plt.figure(figsize=(6,4))
sns.histplot(t_data['age'],kde=True,color='red',bins=12)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Plotting a barplot
plt.figure(figsize=(6,4))
sex_counts = t_data['sex'].value_counts().reset_index()
sns.barplot(data=sex_counts, x='sex', y='count',hue = 'sex', palette='Set1')
plt.title('Gender distribution')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```
