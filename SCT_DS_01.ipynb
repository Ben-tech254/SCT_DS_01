```python
#Importing all the necessary libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

#Loading the dataset
t_data=sns.load_dataset('titanic')

#Plotting a histogram
plt.figure(figsize=(6,4))
sns.histplot(t_data['age'],kde=True,color='red',bins=12)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

#Plotting a barplot
plt.figure(figsize=(6,4))
sex_counts = t_data['sex'].value_counts().reset_index()
sns.barplot(data=sex_counts, x='sex', y='count',hue = 'sex', palette='Set1')
plt.title('Gender distribution')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```
